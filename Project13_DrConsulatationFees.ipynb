```python
def qualification(x):
    b=x.split(' ')
    c=x.split('.')
    d=x.split(',')
    if '-' not in x:
        return b[0]
    elif '.' not in x:
        return c[0]
    elif ',' not in x:
        return d[0]
    else: 
        return x
```

```python
Train_data['Qualification']= Train_data['Qualification'].apply(qualification)
```
